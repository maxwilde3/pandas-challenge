## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,10,14,19,24,29,34,39,100]

group_names = ("<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+")
```

```python
Purchases_df["Age"]
```

```python
pd.cut(Purchases_df["Age"], bins, labels = group_names, include_lowest=True)
```

```python
Purchases_df["Age Summary"] = pd.cut(Purchases_df["Age"], bins, labels = group_names, include_lowest=True)
Purchases_df
```

```python
Purchases_df["Age Summary"].value_counts()
```

```python
Purchases_df["Age Summary"].value_counts(normalize=True)
```
